1.	Reverse a number.

```python
def num(n):
    res=0
    while n>0:
        digit=n%10
        res=res*10+digit
        n//=10
    return res    
print(num(678))        
```

2.	Check if a number is prime.

```python
g = (i*i for i in range(5))
print(g)          # generator object
print(list(g))    # convert to list
```

**Find the smallest number in an array**

```python
def smallest(l):
    min=0
    for i in l:
        if i<min:
            min=i
    return min        
smallest([2,5,1,3,0])  
smallest([8,10,5,7,99])   
```
