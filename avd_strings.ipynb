```python
#1.Write a program to remove all consecutive duplicate characters in a string.
def consecutive(st):
    res=''
    for i in range(len(st)):
        if i==0 or st[i]!=st[i-1]:
            res+=st[i]
    return res
print(consecutive("aaabbccdaa"))        
```

```python
#3.Given two strings, remove all characters from the first string that appear in the second.
def remove_char(st1,st2):
    res=''
    for ch in st1:
        if ch not in st2:
            res+=ch
    return res
print(remove_char('computer','cat'))        
```
