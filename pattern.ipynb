```python
def even(nums):
    for n in nums:
        if n % 2 == 0:
            yield n

a=even([1,2,3,4,5,6])
for i in a:
    print(i)
```

```python
def count(n):
    i = 1
    while i <= n:
        yield i
        i += 1

print(list(count(5)))
```

```python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

print(list(fibonacci(7)))
```

```python
def generate_parentheses(n, open_count=0, close_count=0, current=""):
    
    if len(current) == 2 * n:
        print(current)
        return
    
    if open_count < n:
        generate_parentheses(n, open_count + 1, close_count, current + "(")
    
    if close_count < open_count:
        generate_parentheses(n, open_count, close_count + 1, current + ")")

generate_parentheses(3)
```
