```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
        self.prev=None
class cLL:
    def __init__(self):
        self.head=None
        self.tail=None

    def addfirst(self,data):
        newNode=Node(data)
        if self.head==None and self.tail==None:
            self.head=newNode
            self.tail=newNode
        else:
            newNode.next=self.head
            newNode.prev=self.tail
            self.head.prev=newNode
            self.tail.next=newNode 
            self.head=newNode  

    def addlast(self,data):
        newNode=Node(data)
        if self.head==None and self.tail==None:
            self.head=newNode
            self.tail=newNode
        else:
            newNode.prev=self.tail
            newNode.next=self.head    
            self.tail.next=newNode 
            self.head.prev=newNode   
            self.tail=newNode 

    def addmid(self,pos,data):
        newNode=Node(data)
        if self.head is None:
            print("List is Empty")
            return
        if pos==1:
            self.addfirst(data)
        else:
            n=self.head
            c=1
            while c!=pos-1 and n.next!=self.head:
                n=n.next
                c+=1
            newNode.next=n.next
            newNode.prev=n
            n.next.prev=newNode
            n.next=newNode
            if n==self.tail:
                self.addlast(data) 

    def removefirst(self):
        if self.head==None and self.tail==None:
            print('List is Empty')
            return
        if self.head==self.tail:
            self.head=None
            self.tail=None    
        else:
            self.head=self.head.next   
            self.head.prev=self.tail
            self.tail.next=self.head 

    def removelast(self):
        if self.head==None and self.tail==None:
            print('List is Empty')
            return
        if self.head==self.tail:
            self.head=None
            self.tail=None 
        else:
            self.tail=self.tail.prev  
            self.tail.next=self.head   
            self.head.prev=self.tail

    def removemid(self,pos):    
        if self.head is None:
            print("List is Empty")
            return
        if pos==1:
            self.removefirst()
        else:
            n=self.head
            c=1
            while c<pos and n.next!=self.head:
                n=n.next
                c+=1
            n.prev.next=n.next
            n.next.prev=n.prev
            if n==self.tail:
                self.removelast() 
            if n==self.head:
                print("position out of range")
                return      

    def display(self):
        if self.head is None:
            print('List is empty')
            return
        n=self.head
        print(n.data,end=" ")
        while n.next!=self.head:
            n=n.next
            print("--->",n.data,end=" ")
        print()  

    def displayreverse(self):
        if self.tail is None:
            print('List is empty')
            return    
        n=self.tail
        print(n.data,end=" ")
        while n.prev!=self.tail:
            n=n.prev
            print("--->",n.data,end=" ") 
        print()           
c=cLL()
c.addfirst(88)
c.addfirst(77)
c.display()  
c.displayreverse()
c.addlast(90)
c.addlast(26)
c.display()
c.addmid(4,29292)
c.display()
c.removefirst()
c.display()
c.removelast()
c.display()
c.removemid(3)
c.display()
```
