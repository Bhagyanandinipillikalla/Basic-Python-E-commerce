# count how many even and odd numbers exist in a list

```python
def count_evenodd(l):
    odd_count=0
    even_count=0
    i=0
    while i<=len(l)-1:
        if l[i]%2==0:
            even_count+=1
        else:    
            odd_count+=1
        i+=1
    return even_count,odd_count
li=[int(i) for i in input().split()]    
count_evenodd(li)            
```

# Reverse a list using while loop

```python
def reverse_list(l):
    res=[]
    i=0
    while i<=len(l)-1:
        res=[l[i]]+res
        i+=1
    return res
st=[i for i in input().split()]
reverse_list(st)        
```

# search for a specific element in a list and print its index if found

```python
def specific_ele(l,s):
    i=0
    while i<=len(l)-1:
        if l[i]==s:
            return i
        i+=1
st=[i for i in input().split()]  
st1=input()          
specific_ele(st,st1)           
```

# *count duplicate elements in a list*

```python
def duplicate_element(l):
    res=[]
    i=0
    c=0
    while i<len(l):
        if l[i] not in res:
            res+=[l[i]]
        else:
            c+=1
        i+=1    
    return c
duplicate_element([2,3,4,2,5,3])            

```

# *separate positive and negative numbers from a list and count each*

```python
def separate_pos_neg(l):
    i=0
    pos=[]
    neg=[]
    c=0
    c1=0
    while i<=len(l)-1:
        if l[i]>0:
            pos+=[l[i]]
            c+=1
        if l[i]<0:
            neg+=[l[i]]
            c1+=1
        i+=1  
    return pos,neg,c,c1      
separate_pos_neg([2,-1,9,3,-4,-9,-7,0])            
```

# find the maximum value in a tuple

```python
def max_value(t):
    max=t[0]
    i=0
    while t[i:]:
        if t[i]>max:
            max=t[i]
        i+=1 
    return max
max_value((99,87,56,34))           
```

# count how many elements in a tuple are greater than a given value

```python
def greater_count(t,value):
    c=0
    i=0
    while t[i:]:
        if t[i]>value:
            c+=1
        i+=1
    return c
greater_count((88,98,56,78,45,99),78)            
```

# traverse a dictionary and count the number of key-value pairs

```python
def count(dict):
    keys=list(dict)
    c=0
    i=0
    while i<len(keys):
        c+=1
        i+=1
    return c    
count({1:'a',2:'b',3:'c'})        
```

```python
def anagram(st):
    l=list(st)
    for i in range(len(l)):
        for j in range(len(l)-i-1):
            if l[i]>l[j+1]:
                temp=l[i]
                l[i]=l[j+1]
                l[j+1]=temp
    return "".join(l)            
st1='tone'
st2='note'                
if len(st1)==len(st2) and anagram(st1)==anagram(st2):
    print(True)
else:
    print(False)   
```

```python
def rotate_90(matrix):
    n=len(matrix)
    for i in range(n):
        for j in range(i,n):
            matrix[i][j],matrix[j][i]=matrix[j][i],matrix[i][j]
    for row in matrix:
        row.reverse()
    return matrix
mat=rotate_90([[1,2,3],[4,5,6],[7,8,9]])
print(mat)    
```
