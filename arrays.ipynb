```python
def bigDiff(nums):
    max=nums[0]
    min=nums[0]
    for i in nums:
        if i>max:
            max=i
        if i<min:
            min=i  
    return max-min

print(bigDiff([10,3,5,6]))   
print(bigDiff([7,2,10,9]))
print(bigDiff([2,10,7,2]))           
```

```python
def centerAverage(nums):
    max=nums[0]
    min=nums[0]
    sum=0  
    for i in nums:
        sum+=i
        if i>max:
            max=i
        if i<min:
            min=i
    sum=sum-max-min 
    c=len(nums)-2       
    return sum//c
print(centerAverage([1,2,3,4,100]))  
print(centerAverage([1, 1, 5, 5, 10, 8, 7]))
print(centerAverage([-10, -4, -2, -4, -2, 0]) )         
```

```python
def merge_sorted_arrays(a, b):
    i = j = 0
    result = []

    while i < len(a) and j < len(b):
        if a[i] <= b[j]:
            result.append(a[i])
            i += 1
        else:
            result.append(b[j])
            j += 1

    # add remaining elements
    while i < len(a):
        result.append(a[i])
        i += 1

    while j < len(b):
        result.append(b[j])
        j += 1

    return result
a=[7,5,2,6]
b=[4,1,8,3]
print(merge_sorted_arrays(a, b))
```
