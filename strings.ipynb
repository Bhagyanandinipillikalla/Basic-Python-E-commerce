```python
def second(l):
    max_val = l[0]
    sec_max = None

    for ele in l:
        if ele > max_val:
            sec_max = max_val
            max_val = ele
        elif ele != max_val and (sec_max is None or ele > sec_max):
            sec_max = ele

    return sec_max

print(second([2, 3, 5, 4, 1, 9]))
         
```

```python
def thirdmax(num):
    max1=max2=max3=float('-inf')
    for n in num:
        if n>max1:
            max3=max2
            max2=max1
            max1=n
        elif n>max2:
            max3=max2
            max2=n
        elif n>max3:
            max3=n
    if max3==float('-inf'):
        return max1
    return max3    
print(thirdmax([1,2]))            
print(thirdmax([3,1,2]))        
```
